```python
# Converting the date into datetime format
df["date"] = pd.to_datetime(df.date)
```

```python
# Finding the 5 countries with the highest number of deaths 
df.loc[df["date"] == "2021-03-24"].sort_values(by="total_deaths", ascending=False)[0:5]
```

```python
# Finding the 5 countries with the highest number of Covid-19 cases
df.loc[df["date"] == "2021-03-24"].sort_values(by="total_cases", ascending=False)
```

```python
top5_total = df.loc[df.date == "2021-03-24", ["location", "date", "total_deaths", "total_cases"]]

filt = (top5_total[(top5_total["location"] == "South America") | (top5_total["location"] == "Africa") | (top5_total["location"] == "Asia")
        | (top5_total["location"] == "North America") | (top5_total["location"] == "World") | (top5_total["location"] == "European Union")
         | (top5_total["location"] == "Europe")])

top5_total.drop(filt.index, inplace=True)

top5_total.sort_values(by="total_deaths", ascending=False).head(10)
```

```python
# importing Matplotlib

import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt

# creating a graph of total number of cases
sns.set()

top5_total_deaths_chart = top5_total.loc[top5_total.date == "2021-03-24", ["location", "total_deaths"]].sort_values(by="total_deaths", ascending=False)[0:5]
top5_total_cases_chart = top5_total.loc[top5_total.date == "2021-03-24", ["location", "total_cases"]].sort_values(by="total_cases", ascending=False)[0:5]


fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,7))
ax1.bar(x=top5_total_deaths_chart["location"], height=top5_total_deaths_chart["total_deaths"], facecolor="r", alpha=0.75)
ax1.set_title("Total deaths", fontsize = 20)
ax1.set_ylabel("Deaths", fontsize=15, labelpad=10, )
ax1.grid(ls="--")
ax1.tick_params(labelsize=12)


ax2.bar(x="location", height="total_cases", data = top5_total_cases_chart)
ax2.set_title("Total cases", fontsize=20)
ax2.set_ylabel("Cases", fontsize=15, labelpad=10)
ax2.grid(ls="--")
ax2.yaxis.tick_right()
ax2.yaxis.set_label_position("right")
ax2.tick_params(labelsize=12)


fig.autofmt_xdate()

plt.tight_layout()
```
